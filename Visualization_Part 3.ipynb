```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"HousingData.csv")
```

```python
# Correlation heatmap
```

```python
plt.figure(figsize=(16, 6))

heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap='gist_rainbow_r')

heatmap.set_title('Correlation Heatmap');
```

```python
plt.figure(figsize=(16, 6))

# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(df.corr(), dtype=bool))

heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, mask=mask, annot=True, cmap='gist_rainbow_r')

heatmap.set_title('Correlation Heatmap');
```

```python
# lmplot or linear model plot tells us how does y relates to x. we use lmplot to explore the reltiosnhip between x and y
```

```python
sns.lmplot(data=df2, x='PetalLengthCm', y='PetalWidthCm');  
```

linear relationship between PetalLength and PetalWidth without considering the label

```python
sns.lmplot(data=df2, x='PetalLengthCm', y='PetalWidthCm', hue='Label');
```

linear relationship between PetalLength and PetalWidth when considering the label

```python
sns.lmplot(data=df2, x='SepalLengthCm', y='SepalWidthCm');
```

Negative relationship between Sepal Length and Sepal Width when not considering their lables (class 0, 1 and 2)

```python
sns.lmplot(data=df2, x='SepalLengthCm', y='SepalWidthCm', hue='Label');
```

Positive relationship between Sepal Length and Sepal Width when considering their lables (class 0, 1 and 2)

```python
sns.lmplot(data=df2, x='SepalLengthCm', y='SepalWidthCm', markers=['o', 'x', 'd'], hue='Label');
```

lmplot with different markers

```python
sns.lmplot(data=df2, x='SepalLengthCm', y='SepalWidthCm', markers=['o', 'x', 'd'], hue='Label', col='Label');  # col argument
```

separate lmplot for each class (lablel)

```python
sns.lmplot(data=df2, x='SepalLengthCm', y='SepalWidthCm', markers=['o', 'x', 'd'], hue='Label', col='Label', truncate=False, palette='Set1');  
```
