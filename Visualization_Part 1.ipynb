# What is a violin plot? A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. 




```python
sns.violinplot(data=X, orient='h', palette='OrRd');
```

```python
sns.stripplot(data=X, orient='v', palette='Set2');
```

```python
sns.violinplot(data=X, orient='v', palette='Set2')
sns.stripplot(data=X, orient='v', palette='Set2');
```
