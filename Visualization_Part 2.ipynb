```python
# plot normal distribution (method 1)
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
plt.hist(x, 50, density=True, facecolor='r', alpha=0.75) # density=true  calculates the probability density function for my X


plt.xlabel('IQ')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025,'$\mu=100,\ \sigma=15$')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid(False)
plt.show()
```

```python
# plot normal distribution (method 2)
```

```python
mu = 121 
sigma = 21
x = mu + sigma * np.random.randn(1000)
   
num_bins = 50
   
n, bins, patches = plt.hist(x, num_bins, 
                            density = 1, 
                            color ='green',
                            alpha = 0.7)
   
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))    # y is probability calculated by formula above 
  
plt.plot(bins, y, '-', color ='blue', linewidth=5)
  
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.title('Normal distribution curve');
```

# Plot normal distribution using Scipy library (method 3)

To generate 10000 random numbers from normal distribution mean =0 and variance =1, we use norm.rvs function as

```python
from scipy.stats import norm
data_normal = norm.rvs(size=10000,loc=0,scale=1)
```

```python
ax = sns.distplot(data_normal, kde=True, bins=100, color='skyblue', hist_kws={"linewidth": 25,'alpha':1})

ax.set(xlabel='Normal', ylabel='Frequency');
```

# Generate 10000 random numbers from Poisson random variable with mu = 0.3 and plot them.

```python
from scipy.stats import poisson

data_poisson = poisson.rvs(mu=3, size=10000)
ax = sns.distplot(data_poisson,kde=True,color='red',hist_kws={"linewidth": 15,'alpha':1})

ax.set(xlabel='Poisson', ylabel='Frequency');
```

# generate 10000 from binomial distribution and plot the distribution.

```python
from scipy.stats import binom

data_binom = binom.rvs(n=10,p=0.5,size=10000)
ax = sns.distplot(data_binom, kde=True, color='skyblue', hist_kws={"linewidth": 15,'alpha':1})

ax.set(xlabel='Binomial', ylabel='Frequency');
```
