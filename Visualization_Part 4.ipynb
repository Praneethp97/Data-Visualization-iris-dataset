```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {'C':20, 'C++':15, 'Java':30,'Python':35}
```

```python
type(data)
```

```python
values=data.values()
```

```python
plt.bar(courses, values, color ='maroon', width = 0.4)
plt.xlabel('Grade')
plt.ylabel('No of Students')
plt.title('No of students for each programming class');
```

# Line plot

```python
x=[1,2,3,4,5]    
y=[2,4,6,8,10]
```

Find marker from here:  https://thispointer.com/matplotlib-line-plot-with-markers/#:~:text='%E2%80%9C%20point%20marker,%E2%80%9C'v'%E2%80%9C%20triangle_down%20marker

```python
plt.plot(x,y,'r--P', label='y=2x')   #P  plus    D  Diamond    
plt.legend()
plt.title('test')
plt.xlabel('x')
plt.ylabel('y');
```

```python
x1=[1,2,3,4,5]
y1=[1,4,9,16,25]   
```

```python
plt.plot(x1,y1, 'b:D', label='y=x^2')
plt.legend();
```

```python
plt.plot(x,y,'r--P', label='y=2x')
plt.plot(x1,y1, 'b:D', label='y=x^2')
plt.legend();
```

```python
x=np.linspace(0,20,500)
y1=np.sin(x)
y2=np.cos(x)
```

```python
plt.plot(x,y1, 'r', x, y2, 'b');
```

```python
plt.subplot(211)
plt.plot(x,y1,'r', label='y=sin(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y=isn(x)')


plt.subplot(212)
plt.plot(x,y2, 'b', label='y=cos(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y=cos(x)');
```
