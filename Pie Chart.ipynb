```python
# pie chart 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
l = ['Python', 'Java', 'C++']
s = [ 200    ,  150  , 100 ]
c = ['green', 'red', 'yellow']
```

```python
plt.pie(s, labels=l, colors=c);
```

```python
plt.pie(s, labels=l, colors=c, autopct='%1.1f%%');  # autopct represeens the percantage of each section
```

```python
plt.pie(s, labels=l, colors=c, startangle=90);     # startangle rotates the pie chart
```

```python
plt.pie(s, labels=l, colors=c, explode=(0.1, 0, 0), shadow=True, autopct='%1.1f%%');
```
